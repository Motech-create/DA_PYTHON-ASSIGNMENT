```python
# BMI Calculator
# This program calculates the Body Mass Index of an individual
# Formula: BMI = Weight(kg)/Height(m)**2
import math
```

```python
# Collecting User inputs
weight = float(input('weight'))
height = float(input('height'))

# Calculate BMI
bmi = weight/height**2

#Round to 2 decimal places
bmi = round(bmi, 2)

print ('Your Body Mass Index is:', bmi)

# Categorize BMI using IF statements
if bmi < 18:
    print ('You are Underweight!')
elif bmi > 18  and bmi == 25:
    print ('You are Normal Weight!')
else:
    print ('You are overweight or Obese!')

#BMI Chart Message
msg = '''Body Mass Index Chart:
        Between 18 - 25 = Normal Weight
        Less than (<) 18 = Underweight
        Greater than (>) 25 = Overweight OR Obese'''
print(msg)
```
