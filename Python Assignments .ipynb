```python
# Class 1 Assignments
# Question 1

# BMI Calculator
# This program calculates the Body Mass Index of an individual
# Formula: BMI = Weight(kg)/Height(m)**2
import math
```

```python
# Collecting User inputs
weight = float(input('Enter your weight in kilogram'))
height = float(input('Enter your height in metres'))

# Calculate BMI
bmi = weight/height**2

#Round to 2 decimal places
bmi = round(bmi, 2)

print ('Your Body Mass Index is:', bmi)

# Categorize BMI using IF statements
if bmi < 18:
    print ('You are Underweight!')
elif bmi >= 18  and bmi <=25:
    print ('You are Normal Weight!')
else:
    print ('You are overweight or Obese!')

#BMI Chart Message
msg = '''Body Mass Index Chart:
        Between 18 - 25 = Normal Weight
        Less than (<) 18 = Underweight
        Greater than (>) 25 = Overweight OR Obese'''
print(msg)
```

```python
# Question 2

# Civil Servants Retirement Calculator
# This program helps to calculate and predict the retirement year of a civil servant
```

```python
# Collecting User Inputs
title = input('Enter your title:')
first_name = input('Enter your first name:')
last_name = input('Enter your last name:')
commencement_year_of_service = int(input('Enter the year you commenced civil service:'))
birth_year = int(input('Enter your birth year:'))
current_year = int(input('Enter current year:'))

# Calculations
current_age = current_year - birth_year
retirement_year = birth_year + 60
years_in_service = current_year - commencement_year_of_service
years_left = retirement_year - current_year

# Print Result
print(f"Hello {title} {first_name} {last_name}. You are {current_age} years old and you will be retiring in the year {retirement_year} at the age of 60.")
print(f"You have spent {years_in_service} years of active service. You have {years_left} more years of active service left. Thank you for your service.")
```
